### 3. Functions

A function is a block of reusable code that:
* Takes inputs (called `parameters` or `arguments`)
* performs some calcuations or operations
* Optionally returns an output

Basic Syntax:
```
def function_name(parameters):
    # Code block
    result = parameters * 2
    return result
```
* def -> defines the function
* function_name -> your custom function name
* parameters -> inputs to the function
* return -> gives output to the caller


Suppose we want to convert porosity fracton to percentage


````python
def porosity_percentage(porosity):
    """Convert porosity fraction to percentage."""
    return porosity * 100
````

Note: The section """ """ is called docstring, it is a placeholder where we can add text description of what the function is doing.

````python
# using the function to convert porosity
porosity =  0.25

percentage = porosity_percentage(porosity)

print(f"Porosity percentage: {percentage}%")
````

#### Functions with Multiple Inputs:
In the previous functions, there was only one input. We can also have multiple inputs passed into the function.

Suppose we want to calculate pore volume of a rock layer:
* Pore Volume = Porosity * Layer Thickness * Area

````python
def pore_volume(porosity, thickness_m, area_m2=1):
    
    """Calculate pore volume of a rock layer. """
    volume = porosity * thickness_m * area_m2
    
    return volume
````

````python
# Example
volume = pore_volume(porosity=0.25, thickness_m=120)
print("Pore volume (m³):", volume)
````

````python
# Reusing same function with different inputs
volume = pore_volume(porosity=0.5, thickness_m=50)
print("Pore volume (m³):", volume)
````

#### Functions Returning Multiple Values
Sometimes we may want to return more than one value.

````python
def layer_stats(porosity, thickness):
    """
    Returns pore volume and porosity percentage
    """
    pore_vol = porosity * thickness
    poro_pct = porosity * 100
    return pore_vol, poro_pct
````

````python
# Using the function
pv, pp = layer_stats(0.25, 120)
print("Pore volume:", pv, "m³")
print("Porosity %:", pp)
````

````python
# Using the function
pv, pp = layer_stats(0.5, 50)
print("Pore volume:", pv, "m³")
print("Porosity %:", pp)
````

#### Function with conditional logic 
We can use booleans inside functions. 

Example: classify rock by density

````python
def rock_density_class(density):
    """
    Classify rock based on density (g/cm³)
    """
    if density < 2.4:
        return "Low density"
    elif density <= 2.7:
        return "Medium density"
    else:
        return "High density"

# Test function
print(rock_density_class(2.2))  # Low density
print(rock_density_class(2.5))  # Medium density
print(rock_density_class(3.0))  # High density
````

In the example, we introduced `if..elif..else` statements. These are conditional statements and very important concept in any kind of programming.

We will dive deep into these in the next session.

#### Functions with Default parameters
We can set default values in the parameters of the function. So, when we use the function and don't supply input values, then the functions will run using default values.

````python
def compaction_factor(surface_porosity=0.35, current_porosity=0.25, depth_m=120):
    return (surface_porosity - current_porosity) / depth_m

# Running with default values (no inputs provided, so it uses defaults)
print("Compaction factor:", compaction_factor())

# Can override defaults:
print("Custom compaction:", compaction_factor(current_porosity=0.22, depth_m=150))
````

### 4. Conditions and Conditional Statements

Conditional statements allow to make decisions based on some conditions. Python evaluates the condition and executes a block of code if the condition is True.

#### `if` statement
Basic syntax:
```
if condition:
    # code runs if condition is True
    print("Condition is True")

````python
# Example:
density = 2.5 # g/cm³

if density < 2.4:
    print("Rock is low density") 
````

Here, the condition `density < 2.4` is not met, hence is False, so nothing prints.

`else`statement defines what to do if the condition is False.

````python
if density > 2.7:
    print("Rock is high density")
else:
    print("Rock is not high density")
````

`elif` stagement

This is equivalent to saying `else if` -> checks multiple conditions in order

````python
if density < 2.4:
    print("Low density rock")
elif density <= 2.7:
    print("Medium density rock")
else:
    print("High density rock")
````

#### Logical Operators

We use logical operators to check multiple conditions.
* `and` --> All conditions must be True.
* `or` --> At least one condition is True.
* `not` --> reverse True/False

````python
# Example rock classification 
density = 2.6
porosity = 0.28

# High-quality reservoir rock?
if density <= 2.7 and porosity >= 0.25:
    print("Good reservoir rock")
else:
    print("Not suitable")
````

#### Nested Conditions

We can have conditions inside conditions.

````python
depth_m = 150
porosity = 0.22

if depth_m > 100:
    if porosity > 0.3:
        print("Deep, porous layer")
    else:
        print("Deep, low porosity layer")
else:
    print("Shallow layer")
````

Make sure, the ach if-else statement must have same indentations!

#### Conditional Expresions (Ternary Operator)
We can write a compact if-else statements for a concise clean code

````python
density = 2.5
# --- conditional block ---
if density > 2.7:
    rock_type = "High"
else:
    rock_type = "Medium"
# -------------------------
print(rock_type)
````

````python
# The conditional block in the above can be written more concisely using a conditional expression (ternary operator):
density = 2.5
rock_type = "High" if density > 2.7 else "Medium"  # compact if-else statement
print(rock_type)
````
